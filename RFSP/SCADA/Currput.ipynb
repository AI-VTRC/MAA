```python
import pandas as pd
df = pd.read_csv('SCADA.csv')
df
```

```python
df.ATT_FLAG.unique()
```

```python
import numpy
import collections
import random

df_AIA_add_dict = collections.defaultdict(list)

for i in df.columns[:-1]:

    print(i)
    arr = df[i].values
    mean = numpy.mean(arr, axis=0)
    sd = numpy.std(arr, axis=0)
    for _ in range(500):
        df_AIA_add_dict[i].append(random.uniform(mean+3*sd, mean+4*sd))

            
for _ in range(500):
    df_AIA_add_dict['ATT_FLAG'].append(random.choice(df.ATT_FLAG.unique()))
df_AIA_add = pd.DataFrame(df_AIA_add_dict)

df_AIA_currupted = pd.concat([df,df_AIA_add])
df_AIA_currupted.to_csv('SDACA_currupted_500.csv',index=False)
```
