```python
import pandas as pd
table1 = pd.read_csv('stats_scada.csv',index_col=0)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
table1[table1.columns] = scaler.fit_transform(table1[table1.columns])
table1.fillna(0,inplace = True)

table_AIA = pd.DataFrame()
table_AIA['XAI'] = table1['Kendall'] + table1['ANOVA-F'] + table1['Mutual_info'] + table1['Causality'] 
table_AIA['TAI'] = table1['Kendall'] + table1['ANOVA-F'] + table1['Mutual_info'] + table1['Causality'] 
table_AIA['SAI'] = - table1['Outlier'] - table1['Distribution']
table_AIA['CAI'] = - table1['Outlier'] - table1['Distribution']
table_AIA['FAI'] = - table1['Bias'] - table1['Variance'] 
table_AIA['EAI'] = - table1['Bias'] - table1['Variance']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
table_AIA[table_AIA.columns] = scaler.fit_transform(table_AIA[table_AIA.columns])

df_sum1 = table_AIA.mean().to_frame(name = 'Value')
# scaler = MinMaxScaler()
# df_sum[df_sum.columns] = scaler.fit_transform(df_sum[df_sum.columns])
df_sum1
```
