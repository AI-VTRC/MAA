```python
E,X,F,T,S,C = 16,16,16,16,16,16
```

```python
import pandas as pd

table1 = pd.read_csv('stats_data.csv',index_col=0)
table2 = pd.read_csv('stats_goal.csv',index_col=0)
```

```python
table2.fillna(0,inplace = True)
table2
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
table1[table1.columns] = scaler.fit_transform(table1[table1.columns])
table1.fillna(0,inplace = True)
table1.round(2)
```

```python
table_AIA = pd.DataFrame()
table_AIA['XAI'] = table1['Kendall'] + table1['Chi2'] + table1['ANOVA-F'] + table1['Mutual_info'] + table1['Causality'] 
table_AIA['TAI'] = table1['Kendall'] + table1['Chi2'] + table1['ANOVA-F'] + table1['Mutual_info'] + table1['Causality'] 
table_AIA['SAI'] = - table1['Outlier'] - table1['Distribution']
table_AIA['CAI'] = - table1['Outlier'] - table1['Distribution']
table_AIA['FAI'] = - table1['Bias'] - table1['Variance'] 
table_AIA['EAI'] = - table1['Bias'] - table1['Variance']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
table_AIA[table_AIA.columns] = scaler.fit_transform(table_AIA[table_AIA.columns])
table_AIA
```

```python
table2 = table2[['XAI','TAI','SAI','CAI','FAI','EAI']]
table2
```

```python
table_final = table_AIA * table2
table_final.round(2)
```

```python
df_sum = table_final.mean().to_frame(name = 'Value')
# scaler = MinMaxScaler()
# df_sum[df_sum.columns] = scaler.fit_transform(df_sum[df_sum.columns])
df_sum
```

```python
df_sum = df_sum.T
print(df_sum)
df_sum['XAI'] = df_sum['XAI']*X/100
df_sum['FAI'] = df_sum['FAI']*F/100
df_sum['SAI'] = df_sum['SAI']*S/100
df_sum['CAI'] = df_sum['CAI']*C/100
df_sum['TAI'] = df_sum['TAI']*T/100
df_sum['EAI'] = df_sum['EAI']*E/100
df_sum = df_sum.T
df_sum
```

```python
sum(df_sum.Value)
```
